```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model,preprocessing
from sklearn.model_selection import train_test_split
```

```python
print(sklearn.__version__)
```

```python
data=pd.read_csv("D:\ML mini projects\Car-Data-Set\Car Data Set\car.data")
```

```python
print(data.head(10))
```

```python
# converting the data into numerical form for better computation
# label encoder is used to transform the data to numerical form
le=preprocessing.LabelEncoder()
```

```python
buying=le.fit_transform(list(data['buying']))
maint=le.fit_transform(list(data['maint']))
door=le.fit_transform(list(data['door']))
persons=le.fit_transform(list(data['persons']))
lug_boot=le.fit_transform(list(data['lug_boot']))
safety=le.fit_transform(list(data['safety']))
cls=le.fit_transform(list(data['class']))
# the fit_transform method takes only the list data type and converts it into numerical form
```

```python
label='cls'
X=list(zip(buying,maint,door,persons,lug_boot,safety))
y=list(cls)
# seperating the labels from the features
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(x_train)
# after using the zip function it gives tuples in the list
```

```python
model=KNeighborsClassifier(n_neighbors=7)
# predicting the point by taking the number of neighbors specified in the hyperparameter
model.fit(x_train,y_train)
model.score(x_test,y_test)
# the accuracy of the test after the prediction
```

```python
predict=model.predict(x_test)
# predicting the model
```

```python
names=['unaccurate','accurate','good','vgood']
for i in range(len(predict)):
    print("predicted: ",predict[i],"Data: ",x_test[i],"Actual value: ",y_test[i])
# the predicted and actual data in numerical form
```

```python
for i in range(len(predict)):
    print("predicted: ",names[predict[i]],"Data: ",x_test[i],"Actual value: ",names[y_test[i]])
# the predicted and actual data in real form for readability
```

```python
for i in range(len(predict)):
    n=model.kneighbors([x_test[i]],7,return_distance=True)
#     as it takes only 2D input, enclose it in 2 square brackets
    print("N neighbors: ",n)
# these are the n neighbors(7) with their distances from the point mentioned
```
